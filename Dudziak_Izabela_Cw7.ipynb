```python
import matplotlib.pyplot as plt #instaluję matplotlib
import geopandas as gpd
import os
import glob
```

```python
#lista plików shp
shp_folder=r'/content/drive/MyDrive/Colab Notebooks/dem-rejurb-rejstat-shp'
shp_files = glob.glob(shp_folder + "/*.shp")
len(shp_files)
shp_files
```

```python
#tworzenie GeoDataFrame dla każdego pliku shp
gdfs = {}
for filename in shp_files:
    rejstat = gpd.read_file(filename)
    gdfs[filename] = rejstat
```
